```python
from tkinter import *
import tkinter.font
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)
import numpy as np

root = Tk()
root.title("Grafik Fungsi DFT")
root.geometry("300x600")

changefont = tkinter.font.Font(size=20)
Title = Label(root,text = "TRANSFORMASI FOURIER", font = changefont)
Title.place(x=80, y=10)

def naive_DFT(x):
    X = np.zeros(x.shape,dtype=np.complex128)
    #Jumlah sampel yang diambil
    N = 400
    for m in range(0,N):
        for n in range(0,N):
            X[m] += x[n]*np.exp(-np.pi*2j*m*n/N)
    return X

def plot():
    fig = Figure(figsize = (11, 19),
                dpi = 60)
    #Jumlah sampel yang diambil
    N = 400
    t = np.arange(0.0, 1, 0.0025)
    y = float(A1.get())*np.sin(2*float(w1.get())*np.pi*t)+float(A2.get())*np.cos(2*float(w2.get())*np.pi*t)+float(A3.get())*np.cos(2*float(w3.get())*np.pi*t)
    yf = naive_DFT(y)
    
    plot1 = fig.add_subplot(211)

    plot1.plot(t,y)
    plot1.set_title("Time Domain")
    plot1.set_ylabel('Amplitude')
    plot1.set_xlabel('Time(s)')
    plot1.grid(True)
    
    plot2 = fig.add_subplot(212)
    plot2.plot(np.arange(400), abs(yf))
    plot2.set_title("Frequency Domain")
    plot2.set_ylabel('DFT Amplitude |X(freq)|')
    plot2.set_xlabel('Frequency (Hz)')
    plot2.grid(True)
    
    canvas = FigureCanvasTkAgg(fig,
                            master = root)  
    canvas.draw()
 
    canvas.get_tk_widget().pack()

    toolbar = NavigationToolbar2Tk(canvas,
                                        root)
    toolbar.update()
        
    canvas.get_tk_widget().pack()

# Create an Entry widget
Label(root, font=('Calibri 10')).pack()
A1=Entry(root, width=35)
A1.pack()
A1.place(x=80, y=150)
Label(root, font=('Calibri 10')).pack()
A2=Entry(root, width=35)
A2.pack()
A2.place(x=80, y=220)
Label(root, font=('Calibri 10')).pack()
A3=Entry(root, width=35)
A3.pack()
A3.place(x=80, y=290)
Label(root, font=('Calibri 10')).pack()
w1=Entry(root, width=35)
w1.pack()
w1.place(x=80, y=360)
Label(root, font=('Calibri 10')).pack()
w2=Entry(root, width=35)
w2.pack()
w2.place(x=80, y=430)
Label(root, font=('Calibri 10')).pack()
w3=Entry(root, width=35)
w3.pack()
w3.place(x=80, y=500)

a0 = Label(root, text="Fungsi Sinyal Input = A1*sin(2*w1*pi*t)+A2*cos(2*w2*pi*t)+A3*cos(2*w3*pi*t)")
a0.place(x=80, y=60)
a1 = Label(root, text="A1")
a1.place(x=80, y=130)
a2 = Label(root, text="A2")
a2.place(x=80, y=200)
a3 = Label(root, text="A3")
a3.place(x=80, y=270)
o1 = Label(root, text="w1")
o1.place(x=80, y=340)
o2 = Label(root, text="w2")
o2.place(x=80, y=410)
o3 = Label(root, text="w3")
o3.place(x=80, y=480)

btn = Button(root,text="plot", command=plot).place(x=90, y=570)
root.mainloop()
```
